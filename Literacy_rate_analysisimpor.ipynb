```python
#1.Get top 5 countries with highest adult literacy in 2020.
cursor.execute("""
SELECT Country, adult_literacy_rate_both_sexes
FROM literacy_rates
WHERE year = 2020
ORDER BY adult_literacy_rate_both_sexes DESC
LIMIT 5;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#2.Find countries where female youth literacy < 80%.
cursor.execute("""
SELECT Country, year, youth_literacy_rate_female
FROM literacy_rates
WHERE youth_literacy_rate_female < 80
ORDER BY youth_literacy_rate_female ASC;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#3.Average adult literacy per continent (owid region).
cursor.execute("""
SELECT owid_region,
       AVG(adult_literacy_rate_both_sexes) AS avg_adult_literacy
FROM literacy_rates
WHERE adult_literacy_rate_both_sexes IS NOT NULL
GROUP BY owid_region
ORDER BY avg_adult_literacy DESC;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#illiteracy_population
    # 4. Countries with illiteracy % > 20% in 2000.
cursor.execute("""
    SELECT Country, illiteracy_rate
FROM illiteracy_population
WHERE year = 2000
  AND illiteracy_rate > 20;
  """)

for row in cursor.fetchall():
    print(row)
```

```python
#  5. Trend of illiteracy % for India (2000–2020).
cursor.execute("""
SELECT year, illiteracy_rate
FROM illiteracy_population
WHERE Country = 'India'
  AND year BETWEEN 2000 AND 2020
ORDER BY year;
""")

for row in cursor.fetchall():
    print(row)
```

```python
# 6. Top 10 countries with largest illiterate population in the last year.
cursor.execute("""
SELECT Country, year, illiteracy_rate
FROM illiteracy_population
WHERE year = (SELECT MAX(year) FROM illiteracy_population)
ORDER BY illiteracy_rate DESC
LIMIT 10;
""")

for row in cursor.fetchall():
    print(row)

```

```python
#gdp_schooling
   # 7. Find countries with avg_years_schooling > 7 and gdp_per_capita < 5000.
cursor.execute("""
SELECT Country,
       "average year of education",
       "gdp per capita"
FROM gdp_schooling
WHERE "average year of education" > 7
  AND "gdp per capita" < 5000
ORDER BY "average year of education" DESC;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#8. Rank countries by GDP per schooling for the year 2020.
cursor.execute("""
SELECT Country,
       GDP_per_Schooling_Year,
       RANK() OVER (ORDER BY GDP_per_Schooling_Year DESC)
FROM gdp_schooling
WHERE year = 2020;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#   9. Find global average schooling years per year.
cursor.execute("""
SELECT year,
       AVG("average year of education")
FROM "gdp_schooling"
GROUP BY year
ORDER BY year;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#Join Queries
   #10. List top 10 countries in 2020 with highest GDP per capita but lowest average years of schooling(less than 6).

cursor.execute("""
SELECT Country,
       "gdp per capita",
       "average year of education"
FROM gdp_schooling
WHERE year = 2020
  AND "average year of education" < 6
ORDER BY "gdp per capita" DESC
LIMIT 10;
""")

for row in cursor.fetchall():
    print(row)
```

```python
# 11. Show countries where the illiterate population is high despite having more than 10 average years of schooling.
cursor.execute("""
SELECT g.Country,
       g.year,
       g."average year of education",
       i.illiteracy_rate
FROM gdp_schooling g
JOIN illiteracy_population i
  ON LOWER(TRIM(g.Country)) = LOWER(TRIM(i.Country))
 AND g.year = i.year
WHERE g."average year of education" > 10
  AND i.illiteracy_rate > 20
""")

rows = cursor.fetchall()

if rows:
    for r in rows:
        print(r)
else:
    print("No matching records found")
```

```python
#  12. Compare literacy rates and GDP per capita growth for a selected country over the last 20 years. (country of your choice)
cursor.execute("""
SELECT Country, year, "average year of education"
FROM gdp_schooling
WHERE "average year of education" > 10
ORDER BY "average year of education" DESC;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#  13. Show the difference between youth literacy male and female rates for countries with GDP per capita above $30,000 in 2020.
cursor.execute("""
SELECT l.Country,
       l.year,
       g."gdp per capita",
       l.youth_literacy_rate_male,
       l.youth_literacy_rate_female,
       (l.youth_literacy_rate_male - l.youth_literacy_rate_female)
FROM literacy_rates l
JOIN gdp_schooling g
  ON l.Country = g.Country
 AND l.year = g.year
WHERE l.year = 2020
  AND g."gdp per capita"> 30000
ORDER BY ABS(l.youth_literacy_rate_male - l.youth_literacy_rate_female) DESC;
""")

for row in cursor.fetchall():
    print(row)
```

```python
#12.
query = """
SELECT l.Country,
       l.year,
       g."gdp per capita",
       l.youth_literacy_rate_male,
       l.youth_literacy_rate_female,
       (l.youth_literacy_rate_male - l.youth_literacy_rate_female)
FROM literacy_rates l
JOIN gdp_schooling g
  ON l.Country = g.Country
 AND l.year = g.year
WHERE l.year = 2020
  AND g."gdp per capita"> 30000
ORDER BY ABS(l.youth_literacy_rate_male - l.youth_literacy_rate_female) DESC;
"""

df = pd.read_sql_query(query, conn)

print(df)
df.to_csv("youth_literacy_rate_male,female.csv", index=False)
```
